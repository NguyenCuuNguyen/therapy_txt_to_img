```python
import os
import safetensors.torch

# Define paths to the LoRA weight directories for text encoders
base_checkpoint_dir = "/home/iris/Documents/deep_learning/experiments/custom_finetuned/sdxl/hyperparam_config_0/best_epoch_1"
te1_lora_path = os.path.join(base_checkpoint_dir, "text_encoder_lora")
te2_lora_path = os.path.join(base_checkpoint_dir, "text_encoder_2_lora")

# Function to inspect LoRA weights
def inspect_lora_weights(lora_path, component_name):
    """
    Inspect the LoRA weight file in the given path and print its keys.
    
    Args:
        lora_path (str): Path to the LoRA weight directory.
        component_name (str): Name of the component (e.g., 'Text Encoder 1', 'Text Encoder 2').
    """
    # Check for safetensors file
    safetensors_file = os.path.join(lora_path, "adapter_model.safetensors")
    bin_file = os.path.join(lora_path, "adapter_model.bin")
    
    if os.path.exists(safetensors_file):
        weight_file = safetensors_file
        try:
            weights = safetensors.torch.load_file(weight_file)
            print(f"\n{component_name} LoRA weights ({weight_file}):")
            print("Keys:", list(weights.keys()))
            print(f"Number of parameters: {len(weights)}")
            return weights
        except Exception as e:
            print(f"Error loading {weight_file}: {e}")
            return None
    elif os.path.exists(bin_file):
        print(f"Found .bin file for {component_name}, but .safetensors is preferred. Please convert to .safetensors.")
        return None
    else:
        print(f"No LoRA weight file found in {lora_path}")
        return None

# Inspect weights for both text encoders
te1_weights = inspect_lora_weights(te1_lora_path, "Text Encoder 1")
te2_weights = inspect_lora_weights(te2_lora_path, "Text Encoder 2")

# Optional: Check a few weights for validity (e.g., shape and values)
if te1_weights:
    sample_key = list(te1_weights.keys())[0]
    print(f"Sample weight shape for Text Encoder 1 ({sample_key}): {te1_weights[sample_key].shape}")
if te2_weights:
    sample_key = list(te2_weights.keys())[0]
    print(f"Sample weight shape for Text Encoder 2 ({sample_key}): {te2_weights[sample_key].shape}")
```
